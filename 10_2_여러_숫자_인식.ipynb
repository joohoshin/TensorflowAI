```
import cv2
```

```
image = cv2.imread('/g/My Drive/Tensorflow AI/pi100.png')
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
smaller = cv2.resize(image, dsize=(640,420),interpolation=cv2.INTER_AREA)
cv2_imshow(smaller)
```

```
gray=cv2.cvtColor(smaller, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
print(gray.shape)
print(image.shape)
```

https://opencv-python.readthedocs.io/en/latest/doc/09.imageThresholding/imageThresholding.html

```
thresh = cv2.adaptiveThreshold(gray, 255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,11,2)
cv2_imshow(thresh)
```

https://docs.opencv.org/3.4/df/d0d/tutorial_find_contours.html

```
contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
```

```
contours
```

```
for cnt in contours:
  x,y,w,h = cv2.boundingRect(cnt)
  color = (0,0,255)
  cv2.rectangle(smaller, (x,y), (x+w,y+h), color, 1)
```

```
cv2_imshow(smaller)
```

```
cv2.imwrite('/g/My Drive/Tensorflow AI/numbers_contour.png', image)
```
