# loc
* loc는 인덱스를 가지고 특정 데이터 값을 선택한다. 

```python
import seaborn as sns # 그래픽 라이브러리이며, 데이터셋을 쉽게 불러올 수 있다
# 불러올수 있는 데이터는 여기서 확인이 가능합니다. https://github.com/mwaskom/seaborn-data
```

```python
mpg = sns.load_dataset('mpg') # 차량 무게 및 연비의 데이터입니다. 
```

```python
mpg
```

```python
print(mpg.model_year.max())
print(mpg.model_year.min())
```

```python
mpg.describe()
```

```python
mpg[:10]
```

* loc[행의 조건, 열의 조건]을 통해서 필요한 부분만 선택가능함

```python
mpg.loc[:10]
```

```python
mpg.loc[[1,2,3], ['mpg', 'horsepower']]
```

* 특정 컬럼만 가져오고 행은 전체행이라면 df[:,컬럼]으로 사용한다

```python
mpg.loc[:,'mpg': 'horsepower']
```

* 조건식을 행의 조건 안에 넣을 수 있다

```python
c=mpg.loc[mpg.cylinders==8, :]  # 앞에서 배운 query와 동일한 기능
c
```

```python
mpg.cylinders==8
```

```python
print(type(mpg.cylinders))
print(type(mpg))
```

* iloc는 인덱스의 값과 관계없이 순서대로 가져온다
* index (행의 키값)은 df.index = 리스트 를 넣어주면 된다

```python
c.index = range(10,113)
c
```

```python
c.loc[:11]
```

```python
c.iloc[:10]
```

문자 검색



```python
mpg.loc[mpg['name']=='ford torino',:]
```

```python
# Query 이용 시
mpg.query('name=="ford torino"')
```

*   column.str.contains()를 통해서 일부 단어가 들어간 데이터 검색 가능하다

```python
mpg.loc[mpg.name.str.contains('cadillac')]
```

```python
mpg['name'].str.contains('calillac')
```

```python
mpg.query('name.str.contains("cadillac")', engine='python')  # engine='python' 옵션을 넣어야 오류가 안 남
```
