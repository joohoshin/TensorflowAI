```python
from pycaret.datasets import get_data
dataset = get_data('iris')
```

```python
type(dataset)
```

```python
train = dataset.sample(frac=0.9, random_state=786)
test = dataset.drop(train.index)

train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)

print('Train Dataset for Modeling: ' + str(train.shape))
print('Test Dataset For Predictions: ' + str(test.shape))
```

https://github.com/pycaret/pycaret/blob/master/pycaret/classification.py

```python
from pycaret.classification import *
exp_mclf101 = setup(data = train, target = 'species', session_id=123)
```

```python
exp_mclf101 = setup(data = train, target = 'species', normalize=True, session_id=123)
```

```python
best = compare_models()  # 모델들 Cross Validation으로 비교
```

```python
lr = create_model('lr')  # 기본 파라미터 값으로 생성 됨
```

```python
print(lr)
```

```python
lr_tuned = tune_model(lr)  # 하이퍼파라미터 튜닝
```

```python
lr_tuned
```

```python
pycaret.classification.plot_model(lr_tuned, plot = 'confusion_matrix')  
```

```python
pycaret.classification.plot_model(lr_tuned, plot = 'boundary')  
```
